```python
projects_ = projects_filt[['id','acronym','title','startDate','endDate','totalCost','legalBasis','topics','masterCall','subCall','fundingScheme','objective','grantDoi']]
```

```python
projects_['topics'].tolist()
```

```python
# Define function to assign clusters
def assign_cluster(topic):
    if "CL2" in topic:
        return "Cluster 2"
    elif "CL3" in topic:
        return "Cluster 3"
    elif "CL4" in topic:
        return "Cluster 4"
    elif "CL5" in topic:
        return "Cluster 5"
    elif "CL6" in topic:
        return "Cluster 6"
    elif "HLTH" in topic:
        return "Cluster Health"
    elif "MISS" in topic:
        return "Cluster MISS"
    else:
        return "Other"

# Apply function to create a new column
projects_["cluster"] = projects_["topics"].apply(assign_cluster)
```
